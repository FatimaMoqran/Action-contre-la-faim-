```python
data.head()
```

```python
# calcul des disponibilités en kcal et kg de protéine
data['dispo_alim_kcal'] = data["Disponibilité alimentaire (Kcal/personne/jour)"] * data["Population"] * 365
#je convertis en kg les gramme ouis on multiplie par la pop et par le nbre de jours
data['dispo_alim_kgprot'] = data["Disponibilité de protéines en quantité (g/personne/jour)"] *0.001 * data["Population"] * 365
```
